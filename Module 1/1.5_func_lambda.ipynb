#Functions
Reusable piece of codes. Function parameters can be named or unnamed in Python. Default values can also be used. Consider
the following function.

```python
def say_hello(speaker, person, greeting = "Hello"):
  print(f"{greeting} {person}, this is {speaker}.")

say_hello("amit", "patel")
```

A function is a way to capture code that is commonly executed. Consider the following function that can be used to trim white space from a string capitalize the first letter.

```python
def process_string(str):
  t = str.strip()
  return t[0].upper()+t[1:]
```

```python
str = process_string("  hello   ")
print(f'"{str}"')
```

# Map
The map function takes a list and applies a function to each member of the list and returns a second list that is same as the first list.

```python
l = ['  apple  ', '  pear  ', '  orange  ', '  pine apple   ']
list(map(process_string, l))
```

Map is very similar to **comprehension**. The above can also be done in comprehension.

```python
l2 = [process_string(x) for x in l]
print(l2)
```

#Filter
While a map function always creates a new list of the same size as the original, the filter function creates
a potentially smaller list.

```python
def greater_than_five(x):
  return x>5

l = [1, 20, -5, 34, -2]
l2 = list(filter(greater_than_five, l))
print(l2)
```

#Lambda

Creating a function is sometimes tedious when the same can be done in a line of code. A lambda is essentially an unnamed function.

```python
l3 = list(filter(lambda x: x>5, l))
print(l3)
```

#Reduce
Like maps and filter, reduce also works with  lists. However, the result of list is a single value. Consider if you want to sum the values of a list. the sum is implemented by a lambda.

```python
from functools import reduce

l = [1, 20, 10, -30, -2, 5, 0]
result = reduce(lambda x,y : x+y, l)
print(result)
```
