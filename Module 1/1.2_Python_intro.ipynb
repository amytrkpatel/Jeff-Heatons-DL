# Formatted or interpolated string
You can mix strings and variables in a print statement. This is called formatting string. The variables must be inside curly braces. In Python, this type of string is generally called an f-string. The f-string is denoted by placing an "f" just in front of the opening single or double quote that begins the
string.

````python
a = 10
print(f"the f-value of a is {a}")
````

format() strings:

````python
print("the another way of writing f-value of a plus 5 is {0}".format(a+5))
````

Preferred way is f-string:

````python
a = 5
print(f'Value of a is {a}') #this is the preferred way
print('a is {}'.format(a))
print('a is ' + str(a))
print('a is %d' %(a))
````
